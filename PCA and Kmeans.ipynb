```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data.txt",delim_whitespace=True,header=None,dtype='float32')
```

```python
df.head()
```

```python
X_std = StandardScaler().fit_transform(df)
```

```python
pca = PCA(n_components=20)
principalComponents = pca.fit_transform(X_std)
features = range(pca.n_components_)
print(pca.explained_variance_ratio_)
exp_var_cumul=np.cumsum(pca.explained_variance_ratio_)
plt.plot(features,exp_var_cumul, color='blue')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)
plt.show()
print('A sharp elbow point is visible at x(PCA features)= 2 hence number of components should be 3')
PCA_components = pd.DataFrame(principalComponents)
```

```python
PCA_components
```

```python
cluster_pts = range(1, 10)
inertias = []
for k in cluster_pts:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(PCA_components.iloc[:,:3])
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
plt.plot(cluster_pts, inertias, '-o', color='blue')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(cluster_pts)
plt.show()

print("A sharp elbow point is visible at x(number of clusters,k)= 4 hence number of clusters should be 4")
```

```python
temp=PCA(n_components=3)
data = temp.fit_transform(X_std)

model = KMeans(n_clusters = 4, init = "k-means++")
label=model.fit_predict(data)
centers = np.array(model.cluster_centers_)
```

```python
fig = plt.figure(figsize = (15,12))
ax = plt.axes(projection ="3d")
uniq = np.unique(label)
for i in uniq:
   ax.scatter3D(data[label == i , 0] , data[label == i , 1] ,data[label == i , 2], label = i)
ax.scatter3D(centers[:,0], centers[:,1],centers[:,2], marker="D", color='green')
plt.legend()
plt.show()
```
